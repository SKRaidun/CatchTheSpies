```python
import pandas as pd
import numpy as np

board_data = pd.read_csv('BoardingData.csv', sep=";",
                        usecols=[
                            "PassengerFirstName",
                            "PassengerSecondName",
                            "PassengerLastName",
                            "PassengerBirthDate",
                            "PassengerDocument",
                            "BookingCode",
                            "TicketNumber",
                            "FlightDate",
                            "FlightTime",
                            "FlightNumber",
                            "Destination",
                        ],
                        na_values=["Not presented"],)

board_data["FlightDateTime"] = (board_data["FlightDate"] + ' ' + board_data["FlightTime"]).astype("datetime64[ns]")
board_data["BirthDate"] = board_data["PassengerBirthDate"].astype("datetime64[ns]")
board_data["FlightDate"] =  board_data["FlightDate"].astype("datetime64[ns]")
board_data = board_data.drop(["FlightTime", "PassengerBirthDate"], axis = 1).sort_values(["PassengerFirstName","PassengerSecondName","PassengerLastName"])
board_data["PaxAge"] = (board_data["FlightDate"] - board_data["BirthDate"])//np.timedelta64(365,'D')
board_data = board_data.reset_index(drop=True)
board_data.columns = ["First Name", "Second Name", "Last Name", "TravelDoc", "Code", "e-Ticket", "DepartDate", "Flight", "Destination", "DepartDateTime", "PaxBirthDate", "PaxAge"]

board_data.head()
```

```python
import pandas as pd

flights_data = pd.read_csv('flights_data(PDF).csv')
flights_data.head()
```

```python
import pandas as pd
import numpy as np

forum_data = pd.read_csv('FrequentFlyerForum-Profiles.csv',
                        usecols=[
                            "Last Name",
                            "First Name",
                            "Date",
                            "Flight",
                            "ArrCity",
                            "ArrAirport",
                            "ArrCountry",
                            "DepCity",
                            "DepAirport",
                            "DepCountry",
                            "LoyalProgramm",
                            "LoyalProgNumber"
                        ],
                        )
forum_data_pax_info = forum_data[(forum_data['Last Name'].notnull())]
forum_data_pax_info["DepartDateTime"] = forum_data_pax_info["Date"].astype("datetime64[ns]")
forum_data_pax_info = forum_data_pax_info.drop(["LoyalProgramm", "LoyalProgNumber", 'ArrCountry', 'DepCountry', 'Date'], axis=1).drop_duplicates()
forum_data_pax_info.columns = ['Last Name', 'First Name', 'Flight', 'Destination', 'Dest', 'DepCity', 'From', 'DepartDateTime']
forum_data_pax_info = forum_data_pax_info.reset_index(drop=True)
forum_data_pax_info.head()
```

```python
import pandas as pd

pass_data = pd.read_csv('passenger_info(ZIP).csv',
                       usecols=[
                           "NAME",
                           "FLIGHT",
                           "FROM",
                           "TO",
                           "FROMAIR",
                           "TOAIR",
                           "DATE",
                           "TIME",
                           "E-TICKET",
                       ])
pass_data["First Name"] = pass_data["NAME"].str.split().str[1]
pass_data["Last Name"] = pass_data["NAME"].str.split().str[2]
pass_data = pass_data.drop('NAME', axis=1)
pass_data["DepartDateTime"] = (pass_data["DATE"] + ' ' + pass_data["TIME"]).astype("datetime64[ns]")
pass_data = pass_data.drop(['DATE', 'TIME'], axis=1)
pass_data.columns = ['Flight', 'DepCity', 'Destination', 'From', 'Dest', 'e-Ticket', 'First Name', 'Last Name', 'DepartDateTime']
pass_data.head()
```

```python
import pandas as pd

pass_flights_info = pd.read_csv('Sirena-export-fixed.csv',
                 usecols=[
                     "PaxName",
                     "PaxBirthDate",
                     "DepartDate",
                     "DepartTime",
                     "ArrivalDate",
                     "ArrivalTime",
                     "Flight",
                     "From",
                     "Dest",
                     "Code",
                     "e-Ticket",
                     "TravelDoc",
                     ],
                 na_values=[" "],)

pass_flights_info["Last Name"] = pass_flights_info["PaxName"].str.split().str[0]
pass_flights_info["First Name"] = pass_flights_info["PaxName"].str.split().str[1]
pass_flights_info["Second Name"] = pass_flights_info["PaxName"].str.split().str[2]
pass_flights_info = pass_flights_info.drop('PaxName', axis=1)
pass_flights_info.head()
```

```python
import numpy as np

pass_flights_info["DepartDateTime"] = (pass_flights_info["DepartDate"] + ' ' + pass_flights_info["DepartTime"]).astype("datetime64[ns]")
pass_flights_info["ArrivalDateTime"] = (pass_flights_info["ArrivalDate"] + ' ' + pass_flights_info["ArrivalTime"]).astype("datetime64[ns]")
pass_flights_info["PaxBirthDate"] = pass_flights_info["PaxBirthDate"].astype("datetime64[ns]")
pass_flights_info["DepartDate"] = pass_flights_info["DepartDate"].astype("datetime64[ns]")
pass_flights_info = pass_flights_info.drop(['DepartTime', 'ArrivalTime'], axis=1)
pass_flights_info["PaxAge"] = (pass_flights_info["DepartDate"] - pass_flights_info["PaxBirthDate"])//np.timedelta64(365,'D')
```

```python
import pandas as pd

bonus_programm_data = pd.read_csv('PointzAggregator-AirlinesData.csv',
                                 usecols=[
                                     "first_name",
                                     "last_name",
                                     "card_number",
                                     "bonus_program",
                                     "activity_code",
                                     "activity_date",
                                     "departure",
                                     "arrival",
                                 ])
bonus_programm_data.head()
```

```python
bonus_prog_pax_info = pd.read_csv('PointzAggregator-AirlinesData.csv',
                                 usecols=[
                                     "first_name",
                                     "last_name",
                                     "card_number",
                                 ]).drop_duplicates()

bonus_prog_pax_info.columns = ["FirstName", "LastName", "FF Number"]
bonus_prog_pax_info.head()
```

```python
import pandas as pd

freq_pass_data = pd.read_csv('SkyTeam-Exchange.csv',
                            usecols=[
                                "Date",
                                "Flight",
                                "FF Number",
                                "From",
                                "To",
                            ])

freq_pass_data.head()
```

```python
freq_pass_data = freq_pass_data.merge(bonus_prog_pax_info, on=["FF Number"])
freq_pass_data["DepartDateTime"] = freq_pass_data["Date"].astype("datetime64[ns]")
freq_pass_data = freq_pass_data.drop(['FF Number', 'Date'], axis=1)

freq_pass_data.columns = ['Flight', 'From', 'Dest', 'First Name', 'Last Name', 'DepartDateTime']
freq_pass_data.head()
```

```python
all = pd.concat([board_data, pass_flights_info, pass_data, forum_data_pax_info, freq_pass_data], axis=0).reset_index()
all = all.drop(['DepartDate', 'ArrivalDate', 'index'], axis=1)
all = all.drop_duplicates(subset=['TravelDoc', 'First Name', 'Last Name', 'Flight', 'DepartDateTime', 'e-Ticket'])
all.to_csv('allData.csv', index=False)
```
